## Our Tidy Data

```python
# import libraries

import pandas as pd

# supress warnings
import warnings
warnings.filterwarnings('ignore')
```

## Load CSV Sheets

```python
#  Bealtles Spotify Data


beatles_spotify_csv = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vR8C8FeUUS34zQYIWgMaNQEOOmOqLFw9w8kQsiOmT6SjuKTiLKB8iG9yKWjMJwNzU_ahNYmnhXcBptw/pub?output=csv'


beatles_spotify = pd.read_csv(beatles_spotify_csv)




# and now the CLEAN BILLBOARD DATA


beatles_billboard_clean_csv = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTvOhZSzoa8kwkmYOzwuAGq3Piaujeabu41wVgAhSPfS0IONz2zi_nG5Km_5GO8B8P6tor0r8CacyNF/pub?output=csv'


beatles_billboard = pd.read_csv(beatles_billboard_clean_csv)


# and perhaps OUR Cleaned Metadata too!


```

### Let's Tidy The Data by Exploding the Genres, so we have ONE Genre per Row

```python


# split the long strings into a list of strings in each cell
beatles_billboard['Genre'] = beatles_billboard['Genre'].str.split(', ')
```

### Now explode on Genre column

```python

# explode the data on the genre column, and save that as a new df
beatles_billboard_exploded = beatles_billboard.explode('Genre').reset_index(drop=True)
beatles_billboard_exploded
```

## Join the DataFrames

```python
combined_beatles_data = pd.merge(beatles_spotify, beatles_billboard_exploded, left_on='song', right_on='Title', how='inner')
combined_beatles_data
```

```python
# drop redundant columns
combined_beatles_data = combined_beatles_data.drop(columns=['Title', 'Year', 'Duration', 'Other.releases'])
combined_beatles_data
```

### Now we have other problems with Genre, since the individual tags need to be cleaned and regularized

```python
# clean individual problems in the exploded data with str.replace() and str.strip()
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.strip('[')
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace('pop/rock', 'pop rock')
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace('r&b', 'rhythm and blues')
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace('rock and roll', 'rock')
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace('experimental music', 'experimental')
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace("children's music", "children's")
combined_beatles_data['Genre'] = combined_beatles_data['Genre'].str.replace("stage&screen", "stage and screen")
```

### What about the original Spotify Audio Features?

We could reshape the Audio Features with pd.melt()

- For example we might want ONE variable per column for the audio features, and then a 'value' column with the values for each feature.  



```python
# Choose our ID Vars, or identifiers - the column to keep fixed
id_vars = ['song']

# Choose our value_vars, or columns to "unpivot" into a single "variable" column.
value_vars = ['danceability', 'energy', 'speechiness', 'acousticness', 'liveness', 'valence']

# Using melt
melted_feature_data = pd.melt(
    beatles_spotify, # Dataset
    id_vars=id_vars, # Pivot
    value_vars=value_vars # Values
).sort_values('song').reset_index(drop=True) # Sorting

melted_feature_data
```
